# Data Visualization: Wrangled Medical Supplier Dataset (main)

This notebook builds on the wrangled dataset and creates clear, purposeful visualizations to explore the structure and distribution of the data.

## Load the Wrangled Dataset

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the wrangled dataset
df = pd.read_csv("../data/processed/medical_suppliers_wrangled.csv")
df.shape
```

## Providers per Region

```python
region_counts = df['region'].value_counts()
region_counts.plot(kind='bar', color='skyblue', edgecolor='black')
plt.title("Number of Providers by Region")
plt.xlabel("Region")
plt.ylabel("Count")
plt.xticks(rotation=0)
plt.tight_layout()
plt.show()
```

## Account Age Distribution

```python
df['account_age_years'].dropna().plot(kind='hist', bins=20, color='lightgreen', edgecolor='black')
plt.title("Account Age Distribution")
plt.xlabel("Years")
plt.ylabel("Frequency")
plt.tight_layout()
plt.show()
```

## Price Distribution (Boxplot)

```python
df.boxplot(column='price', vert=False)
plt.title("Price Distribution (Simulated)")
plt.xlabel("Price")
plt.tight_layout()
plt.show()
```

## Top 10 Specialities

```python
plt.figure(figsize=(10, 6))

df['first_speciality'].value_counts().head(10).plot(
    kind='bar',
    color='salmon',
    edgecolor='black'
)

plt.title("Top 10 Types of Supplies")
plt.xlabel("Speciality")
plt.ylabel("Count")
plt.xticks(rotation=90)
plt.tight_layout()
plt.show()
```

## Summary
- These plots are for quick insights into supplier geography, tenure, price spread, and medical focus areas.
- They complement the structured cleaning and wrangling steps shown in previous notebooks.
